### **1. Drop duplicates**

```python
df = df.drop_duplicates()
df
```

## **2. Drop not useful columns**

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

## **3. Last Name's cleaning**

```python
## Remove not wanted chars and blank spaces
df['Last_Name'] = df['Last_Name'].str.lstrip('...').str.lstrip('/').str.rstrip('_').str.strip()
df

## Another way to do the same thing
#df['Last_Name'] = df['Last_Name'].strip('123._/')
```

## **4. Phone Number's cleaning**

```python
## Some cleaning before
df['Phone_Number'] = df['Phone_Number'].str.replace('-', "").str.replace('|', "").str.replace('/', "")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# More cleaning
df['Phone_Number'] = df['Phone_Number'].replace('nan--', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('Na--', '')
df
```

##**5. Address cleaning**

```python
df.head()
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)
```

```python
df = df.drop(columns='Address')
```

```python
df
```

##**6. Paying Customer's and DoNotContact's cleaning**

> Bloc con sangra



```python
df['Paying Customer'] = df['Paying Customer'].replace('Y', 'Yes')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('N', 'No')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N', 'No')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Y', 'Yes')
df
```

##**6. General dataframe cleaning**

```python
df= df.replace('N/a', '')
```

```python
df = df.fillna('')
df
```
